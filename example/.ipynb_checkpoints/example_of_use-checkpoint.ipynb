```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
```

```python
path_train = 'data/base_train.csv'
path_test = 'data/base_test.csv'
df_train = pd.read_csv(path_train) #get train set as a dataframe
df_test = pd.read_csv(path_test) #get test set as a dataframe
df_train #show dataframe training-set
```

```python
df_test #show dataframe test-set
```

```python
#separate xi and y features as a matrix/array
xi_train = df_train.iloc[:, 1:].values
xi_test = df_test.iloc[:, 1:].values
y_train = df_train.iloc[:, 0].values
y_test = df_test.iloc[:, 0].values
xi_train #show xi_train matrix
```

```python
y_test #show y_test array
```

```python
#traning a regression model (sklearn)
rlr = RandomForestRegressor(n_estimators=150)
rlr.fit(xi_train, y_train)
```

## Work with pyADA

```python
from pyADA import Smetrics, Similarity, ApplicabilityDomain
```

```python
# Internal documentation
help(Similarity)
```

```python
help(ApplicabilityDomain)
```

```python
AD = ApplicabilityDomain(verbose=True)
```

```python
#Analysis of similarities between the structures of the test-set and training-set 
sims = AD.analyze_similarity(base_test=xi_test, base_train=xi_train,
                             similarity_metric='tanimoto')
sims
```

```python
#All similarities values
AD.similarities_table_
```

```python
#scan of similarities to found threshold
result = AD.fit(model_sklearn=rlr, base_test=xi_test, base_train=xi_train,
                y_true=y_test, threshold_step=(0, 1, 0.1), threshold_reference='max',
                similarity_metric='tanimoto', metric_error='rmse')
```

```python
result
# EX: For threshold 0.4 RMSE = 5351739296.605793 and Test-samples <= 0.4 similarities = [ 15,  18,  99, 100, 113, 114, 115, 121, 125, 135]
```

```python
# view result as a dataframe
results_df = pd.DataFrame(result).T
results_df.columns = ['RMSE', 'Samples_LT_Threshold']
results_df
```
